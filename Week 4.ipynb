```python
import pandas as pd
from numpy.random import seed
from numpy.random import rand
from numpy.random import randint
import numpy as np

seed(42)

df = pd.DataFrame({'Weather' : ['cold', 'hot', 'cold', 'hot',
   'cold', 'hot', 'cold'],
   'Food' : ['soup', 'soup', 'icecream', 'chocolate',
   'icecream', 'icecream', 'soup'],
   'Price' : 10 * rand(7), 
    'Number' : randint(1, 9)})

print(df)
```

```python
# group by 

df_group_weather = df.groupby("Weather")

print(df_group_weather)
```

```python
for grp_name, grp_item in df_group_weather:
    print(f"Group: {grp_name}")
    print(grp_item)
```

```python
# group first row
df_group_weather.first()
```

```python
# group last row
df_group_weather.last()
```

```python
# group mean
df_group_weather.mean()
```

```python
# Group on Food
df_group_food = df.groupby(["Weather","Food"])
df_group_food
```

```python
df_group_food.agg([np.mean, np.median])
```
